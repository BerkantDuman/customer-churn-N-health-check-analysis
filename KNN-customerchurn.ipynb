```python
import pandas as pd
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
from sklearn import preprocessing
%matplotlib inline
```

```python
dataFM = pd.read_excel("customers2.xlsx", engine="openpyxl")
```

```python
label_encoder = preprocessing.LabelEncoder() # Karakter olanları encodelarayark numeric hale getiriyoruz
dataFM['state'] = label_encoder.fit_transform(dataFM['state'])
dataFM['international plan'] = label_encoder.fit_transform(dataFM['international plan'])
dataFM['voice mail plan'] = label_encoder.fit_transform(dataFM['voice mail plan'])
```

```python
dataFM.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import neighbors, datasets, preprocessing
from sklearn.metrics import accuracy_score
```

```python
dataFM = dataFM.drop("phone number", axis=1)
x = dataFM.drop("churn", axis=1).values
y = dataFM["churn"].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3 , random_state = 0) # test ve eğitim verisi olarak ayırdık
scaler = preprocessing.StandardScaler().fit(x_train) # Veriler ölçeklendi
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
knn = neighbors.KNeighborsClassifier(n_neighbors=5) #KNN modeli oluşturuldu
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
accuracy_score(y_test, y_pred) # başarı yüzdesi
```

```python
compare = pd.DataFrame(y_test, columns = ["Gerçek Y"])
compare2 = pd.DataFrame(y_pred, columns = ["Tahmin Y"])
tahminDataFrame = pd.concat([compare, compare2], axis = 1)
```
